# Visualising CfD Auction Clearing Prices

<br>

### Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

<br>

### Helper Functions

```python
def hide_spines(ax, positions=['top', 'right']):
    """
    Pass a matplotlib axis and list of positions with spines to be removed
    
    args:
        ax:          Matplotlib axis object
        positions:   Python list e.g. ['top', 'bottom']
    """
    assert isinstance(positions, list), 'Position must be passed as a list '
    
    for position in positions:
        ax.spines[position].set_visible(False)
```

<br>

### Reading in the Data

```python
df_CfDs = pd.read_csv('Data/CfD_rounds.csv')
df_CfDs.head()
```

<br>

### Cleaning the Data

```python
df_CfDs = df_CfDs.rename(columns={'Capacity\n(MW)':'Capacity (MW)'})

cols_2_convert = ['Capacity (MW)', 'Homes Powered']

for col in cols_2_convert:
    df_CfDs[col] = (df_CfDs
                    [col]
                    .str.replace(',', '')
                    .astype(float)
                   )
    
df_CfDs.head()
```

<br>

### Creating the Visualisation

```python
s_commissioned = (df_CfDs['Year End'] + df_CfDs['Year Start'])/2
cfd_rounds = range(1, 4)

## Plotting
cmap = 'winter'
palette = sns.color_palette(cmap, len(cfd_rounds))
date_start = 2011.5

fig, ax = plt.subplots(dpi=750, figsize=(7, 4))

sctr_plt = dict()

for cfd_round in cfd_rounds:
    s_round_bool = df_CfDs['Round']==cfd_round
    clearing_price = df_CfDs[s_round_bool]['Strike Price (£/MWh)'].max()
    
    ax.plot([date_start, 2040], [clearing_price, clearing_price], '-.', color=palette[cfd_round-1], alpha=0.25)
    sctr_plt[cfd_round] = ax.scatter(s_commissioned[s_round_bool], 
                                     df_CfDs[s_round_bool]['Strike Price (£/MWh)'], 
                                     s=df_CfDs[s_round_bool]['Capacity (MW)'], 
                                     alpha=0.5, color=palette[cfd_round-1])#, color='deepskyblue')
    
    ax.text(date_start+0.3, clearing_price+2, f'Round {cfd_round}', color=palette[cfd_round-1], weight='bold')

ax.set_xlim(date_start, 2040)
ax.set_ylim(0, 140)
ax.set_xlabel('Commissioning Year')
ax.set_ylabel('CfD Strike Price (£2012/MWh)')

leg = ax.legend(*sctr_plt[cfd_round].legend_elements("sizes", num=6), labelspacing=3, handletextpad=1.2, bbox_to_anchor=(1.3, 1.05), frameon=False, title='Capacity (MW)')
leg.get_title().set_position((-10, 0))
for marker in leg.legendHandles:
    marker.set_color('deepskyblue')

hide_spines(ax)

fig.tight_layout()
fig.savefig('img/CfD_clearing_prices.png')
```
